### Cantidad de vacunas aplicadas de cada compañía
c. Determinar la cantidad de vacunas aplicadas de cada compañía (en base a como lo reporta cada país en la columna vaccines)

#NOTA:
En este caso, la segunda parte de la instrucción ya que la columna *''vaccines''* es referente a un subconjunto de datos denominado *''country_vaccinations_by_manufacturer.csv''*  de manera que se presentan ambos casos aplicables a la consigna para su revisión. 

```python
vacCO = df2.groupby(['vaccine'], as_index=False)['total_vaccinations'].sum()
vacCO['total_vaccinations'] = vacCO['total_vaccinations'].apply(lambda x: "{:,.0f}".format(x))

print('Vacunas aplicadas de cada compañia:')
display(vacCO.head())

vacCO2 = df2.groupby(['location','vaccine'], as_index=False)['total_vaccinations'].sum()
vacCO2['total_vaccinations'] = vacCO2['total_vaccinations'].apply(lambda x: "{:,.0f}".format(x))

print('Vacunas aplicadas de cada compañia en relación al país')
display(vacCO2.head())
```

### Dataframe nuevo llamado conCantidad
i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac)<br> con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter / )

#NOTA:
De acuerdo con la ultima actualización del set de datos <a href="https://www.kaggle.com/datasets/gpreda/covid-world-vaccination-progress">COVID-19 World Vaccination Progress</a> *(versión 249 - 2022/03/31)* la columna *''vaccines''* está ligada al sub conjunto de datos llamado *''country_vaccinations_by_manufacturer.csv''* el cual contiene la información de cada vacuna y el caracter "/" no define una separación entre vacunas. 

Por lo que la consigna para este ejercicio no sólo es confusa, sino poco adecuada a los datos que tenemos disponibles. 

Sin embargo, mi propuesta de extracción para cumplir con la visualñisación solicitada es:

* Tomar el subconjunto de datos de la columna *''vaccines''*
* Ordenarlos de manera ascendente con la lógica : locación, fecha, vacuna
* Agrupar y obtener la diferencia de las vacunas de *''hoy''* y las de el día anterior para así obtener las vacunas aplicadas cada día por cada compañía, 





```python
conCantidad = df2.copy()

conCantidad = conCantidad.sort_values(by=['location', 'date', 'vaccine'])
conCantidad['canVac'] = conCantidad.groupby(['location', 'vaccine'])['total_vaccinations'].diff()
conCantidad['canVac'] = conCantidad['canVac'].fillna(0)
conCantidad['canVac'] = conCantidad['canVac'].apply(lambda x: "{:,.0f}".format(x))

conCantidad
```

### Dataframe llamado pfizer
k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

### Almacenar los dataframes
l. Almacenar los dataframes generados (conDiferencias,conCantidad,antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx,<br> donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter

### NOTA: 
En este caso el ciclo se asegura de retirar los hipervínculos ya que el formato xlsx, a diferencia de csv, los almacena como objetos y el numero es mucho más limitado que el texto plano. 


```python

workbook = Workbook()
sheet_names = ['conDiferencias','conCantidad', 'antes20', 'pfizer']


for sheet_name, df in zip(sheet_names, [conDiferencias,conCantidad, antes20, pfizer]):
    worksheet = workbook.create_sheet(title=sheet_name)
    df['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d')
    rows = dataframe_to_rows(df, index=False)

    for r_idx, row in enumerate(rows, start=1):
        for c_idx, value in enumerate(row, start=1):
            worksheet.cell(row=r_idx, column=c_idx, value=value)

    for cell in worksheet['A']:
        if cell.hyperlink is not None:
            cell.hyperlink = None

workbook.remove(workbook['Sheet'])
workbook.save('resultadosReto.xlsx')
```
